```python
# 1. Load the "--.csv" dataset using pandas.
# 2. Perform the following data-cleaning tasks:
    
    
# Apply all the basic pandas function to get to know about your
# dataset.

import pandas as pd
df = pd.read_csv('customerdemographic.csv')

df.head()                 #View the first few rows of the dataset
df.tail()                 #View the last few rows of the dataset
df.info()                  #Get an overview of the DataFrame structure and data types of each column
df.describe()               #Get descriptive statistics for numerical columns in the DataFrame
df.shape                  #Check the dimensions of the DataFrame (number of rows and columns)
df.columns                     #Get the names of all the columns in the DataFrame
df.dtypes                  # this function wil show the data types of every column in dataset
df.count()                      #Count the number of non-null values in each column
df.isnull().sum()                #Check for any missing values in the DataFrame
df.dropna()                     #Drop any rows with missing values
df.sort_values('past_3_years_bike_related_purchases')       #Sort the DataFrame by a specific column
df['DOB'] = pd.to_datetime(df['DOB'])                       #converting the 'DOB' column to a datetime data type
df = df.rename(columns={'DOB': 'Date of Birth'})            #rename column
select_column1 = 'past_3_years_bike_related_purchases'
outliers = df[(df[select_column1] < df[select_column1].mean() - 4 * df[select_column1].std()) | (df[select_column1] < df[select_column1].mean() + 4 * df[select_column1].std())]
print('Outliers: \n')               # this function will print the outliers in the mentioned column
outliers
```

```python
# Apply replace in the gender function to give a standard
# sturcture to each category.

import pandas as pd
df = pd.read_csv('customerdemographic.csv')

df = df.rename(columns={'gender': 'Gender'})  
df['Gender'].unique()
gender_mapping = {'F': 'Female', 'M': 'Male', 'U': 'Unknown', 'Femal': 'Female'} 
df['Gender'] = df['Gender'].replace(gender_mapping)
df['Gender'].unique()
df.head()


# These functions will show the updated unique values in the "gender" column with the standardized structure.

# By following these steps, we can apply the replace function to give a standard structure to each 
# gender category in the "gender" column of we dataset.
```

```python
# Add a new column Age by deriving the Date of Birth of the
# candidate from DoB Column.

import pandas as pd
df = pd.read_csv('customerdemographic.csv')

# Converting the "DOB" column to a datetime format 
df['DOB'] = pd.to_datetime(df['DOB'])

# To Calculate the current date or a reference date for calculating the age
reference_date = pd.to_datetime('2023-06-24')  

# To Calculate the age based on the exact date of birth and the reference date
df['Age'] = (reference_date - df['DOB']).dt.days // 365



df[['DOB', 'Age']]


# These functions will show the updated unique values in the "gender" column with the standardized structure.

# By following these steps, we can apply the replace function to give a standard structure to each 
# gender category in the "gender" column of we dataset.
```
